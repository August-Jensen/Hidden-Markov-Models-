# Hamilton Algorithm for Comparing AR(1) with Statsmodels Markov Autoregression

This code aims to establish a Hamilton algorithm for an AR(1) model, with n states, and then comparing Filters performance with the one from Statsmodels.


```python
# insert code here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #pyplot is used to plot the data
import scipy.optimize as opt #used to numerically optimize
from datetime import time, timedelta, datetime
import importlib

# Import Dataframe
df = pd.read_csv('SP500.csv', sep=';')
df['pct'] = df['Return'] * 100
df['cum'] = df['Return'].cumsum()
df
```

## Hamilton Filter


I first create a Model, that can calculate the likelihood, fit the model, plot filtered and smoothed probabilities, the estimated model, the observed data, and the residuals, and provide tesst statistics.  

```python
import Hamilton
importlib.reload(Hamilton)
# model = Hamilton.fit(df['pct'], n_states=2)
# log_likelihood_contribution = model.log_likelihood_contribution(t=1, state=0)



# Step 1: Load your data (this is just a placeholder; replace it with your data)
# y = np.random.randn(100)  # Replace with your actual time series data
y = df['pct']

# Step 2: Initialize the HamiltonFilter model
# These are arbitrary initial values; replace them with your actual initial guesses
initial_sigma2 = np.var(y)
initial_params = np.array([0.95, 0.95, np.sqrt(2 * initial_sigma2), np.sqrt(0.5 * initial_sigma2)])
bounds = [(0.001, 0.9999), (0.001, 0.9999), (0.01, None), (0.01, None)]

model = Hamilton.HamiltonFilter(n_states=2, sigma2=initial_params[2:4] ** 2, p00=initial_params[0], p11=initial_params[1])

# Step 3: Optimize parameters
estimated_params, standard_errors = model.optimize_parameters(y, initial_params, bounds)

# Step 4: Run the filter
xi_10, xi_11, xi_1T = model.forward_backward_filter(y, model.transition_matrix, model.sigma2)

# Compute filtered volatility
vol = xi_11[0, :] * model.sigma2[0] + (1 - xi_11[0, :]) * model.sigma2[1]

# Step 5: Plot the results
model.plot_results(y, vol)
model.plot_probabilities(xi_10, xi_11, xi_1T, len(y))

```
